## Описание проекта

### Описание данных

В файле `games.csv` собраны исторические данные из открытых источников о продажах игр, оценки пользователей и экспертов, жанры и платформы

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB 

ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Возможные значения рейтинга (источники - [ESRB](https://www.esrb.org/ratings-guide/), [Википедия](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board))

- `EC` (**Early Childhood**) - «_Для детей младшего возраста_»: для детей от 3 лет
- `E` (**Everyone**) - «_Для всех_»: для возрастной категории от 6 лет
- `E10+` (**Everyone 10+**) - «_Для всех от 10 лет и старше_»
- `T` (**Teen**) - «_Подросткам_»: для лиц от 13 лет
- `M` (**Mature**) - «_Для взрослых_»: не подходит для подростков младше 17 лет
- `AO` (**Adults Only 18+**) - «_Только для взрослых_»: для взрослых старше 18 лет
- `RP` (**Rating Pending**) - «_Рейтинг ожидается_»: продукт ожидает присвоения рейтинга

Данные за 2016 год могут быть неполными.

### Исследование и заполнение пропусков


Проанализируем структуру пропусков в таблице.

```python
# check and calculate all skips
data.isna().sum()
```

```python
# пропущенные значения бары

def pass_value_barh(df):
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by = 'space', ascending = True)
            .plot(kind = 'barh', figsize = (19,6), rot = 0, legend = False, fontsize = 16)
            .set_title('Доля пропусков в данных' + "\n", fontsize = 22, color = 'SteelBlue')    
        );    
    except:
        print('пропусков не осталось :) ')
```

```python
pass_value_barh(data)
```

Большая часть пропусков в столбцах: `'user_score'`, `'critic_score'`, `'rating'`, которые содержат в себе оценки пользователей, критиков и рейтинг от организации ESRB. Возможная причина пропусков - какие-либо регионы не оценивали игры по различным параметрам или не присваивали рейтинг играм определенных регионов, так как не используется рейтинг от организации ESRB.

Коэффициент Пирсона отражает, что связь между продажами и оценками критиков очень слабая. Связь отрицательная.

```python
# Определим функцию, которая будет отрисовывать диаграммы рассеивания и считать корреляции
platform_top_6 = ['3DS', 'PC', 'PS4', 'PSV', 'WiiU', 'XOne']
def platform_matrix(name_of_platform):
    platform = data[(data['platform']==name_of_platform) & (data['critic_score'] >= 0) & (data['year_of_release'] >= 2013) & (data['user_score'] >= 0)]
    fig, ax = plt.subplots(1 ,2, figsize=(15,5))
    sns.scatterplot(x='user_score', y='total_sales', data=platform, ax=ax[0])
    sns.scatterplot(x='critic_score', y='total_sales', data=platform, ax=ax[1])
    fig.suptitle(name_of_platform, fontsize=15)
    ax[0].set(xlabel='Оценка пользователей')
    ax[1].set(xlabel='Оценка критиков')
    ax[0].set(ylabel='Количество продаж')
    ax[1].set(ylabel='Количество продаж')
    plt.show()
    
    correlation = platform['user_score'].corr(platform['total_sales'])
    critic_correlation = platform['critic_score'].corr(platform['total_sales'])
    if 0.1 >= critic_correlation >= -0.2:
        print('Корреляция между оценками критиков и игровой платформой ', name_of_platform, ': Слабая - отрицательная', critic_correlation)
    if 0.3 >= critic_correlation >= 0.1:
        print('Корреляция между оценками критиков и игровой платформой ', name_of_platform, ': Слабая', critic_correlation)
    if 0.5 >= critic_correlation >= 0.3:
        print('Корреляция между оценками критиков и игровой платформой ', name_of_platform, ': Невысокая', critic_correlation)
    if 0.7 >= critic_correlation >= 0.5:
        print('Корреляция между оценками критиков и игровой платформой ', name_of_platform, ': Высокая', critic_correlation)
    if 0.9 >= critic_correlation >= 0.7:
        print('Корреляция между оценками критиков и игровой платформой ', name_of_platform, ': Весьма высокая', critic_correlation)
    if 1 >= critic_correlation >= 0.9:
        print('Корреляция между оценками критиков и игровой платформой ', name_of_platform, ': Сильная', critic_correlation)
    
    if 0.1 >= correlation >= -0.2:
        print('Корреляция между оценками пользователей и продажами ', name_of_platform, ': Слабая - отрицательная', correlation)
    if 0.3 >= correlation >= 0.1:
        print('Корреляция между оценками пользователей и продажами ', name_of_platform, ': Слабая', correlation)
    if 0.5 >= correlation >= 0.3:
        print('Корреляция между оценками пользователей и продажами ', name_of_platform, ': Невысокая', correlation)
    if 0.7 >= correlation >= 0.5:
        print('Корреляция между оценками пользователей и продажами ', name_of_platform, ': Высокая', correlation)
    if 0.9 >= correlation >= 0.7:
        print('Корреляция между оценками пользователей и продажами ', name_of_platform, ': Весьма высокая', correlation)
    if 1 >= correlation >= 0.9:
        print('Корреляция между оценками пользователей и продажами ', name_of_platform, ': Сильная', correlation)
    
    print('\n')


```

```python
# Используя цикл, построим диаграммы рассеивания и рассчитаем корреляцию
for platform in platform_top_6:
     platform_matrix(platform)
```

На основании расчетов и графиков можно отметить, что невысокая корреляция между оценками критиков и продажами присутствует у платформ: 3DS, PS4, XOne, WiiU. Для остальных платформ - PC, PSV связь слабая.

Корреляция невысокая между оценками пользователей и продажами присутствует у платформ: WiiU. Для остальных платформ связь слабая, и часто отрицательная.

## Проверка гипотез

###  Пользовательские рейтинги платформ Xbox One и PC

Для проверки гипотезы о равенстве средних пользовательских рейтингов платформ Xbox One и PC рассматриваем 2 выборки. Рассчитаем основные статистики для них.

```python
# рассчитаем основные статистики
platform_stat = ['XOne', 'PC']
stats_data_1 = data.query('platform in @platform_stat and year_of_release >=2013 and user_score >=0') \
.pivot_table(index=['platform'],\
            values=['user_score'],\
            aggfunc=['mean', 'std', 'median']\
).round(2).reset_index()
stats_data_1
```

Из таблицы видно, что среднее квадратическое отклонение различно для данных выборок.

Рассмотрим гипотезы:

Нулевая:

H_0: Средние пользовательские рейтинги платформы Xbox One  = Средние пользовательские рейтинги платформы PC
`
Альтернативная:

H_a: Средние пользовательские рейтинги платформы Xbox One  ≠ Средние пользовательские рейтинги платформы PC




```python
alpha=0.05 # Статистическая значимость 
```

Так как мы рассматриваем гипотезу о равенстве средних, то будем использовать  метод для проверки гипотезы scipy.stats.ttest_ind (array1, array2, equal_var)

```python
# подготовим выборки
rating_xone = data[(data['platform'] == "XOne") & (data['year_of_release'] >=2013) & (data['user_score'] >=0)]['user_score']
rating_pc = data[(data['platform'] == "PC") & (data['year_of_release'] >=2013) & (data['user_score'] >=0)]['user_score']

# вызов метода для проверки гипотезы
results = st.ttest_ind(rating_xone, rating_pc, equal_var=False) 

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вывод: Значение p-value приблизительно 15%. Таким образом, не получилось опровергнуть Нулевую гипотезу. То есть, с вероятностью в 15% можно получить различность рейтингов двух платформ

### Пользовательские рейтинги жанров Action и Sports

Для проверки гипотезы о различии средних пользовательских рейтингов жанров Action и Sports рассматриваем 2 выборки. Рассчитаем основные статистики для них.

```python
# рассчитаем основные статистики
genre_stat = ['Action', 'Sports']
stats_data_2 = data.query('genre in @genre_stat and year_of_release >=2013 and user_score >=0') \
.pivot_table(index=['genre'],\
            values=['user_score'],\
            aggfunc=['mean', 'std', 'median']\
).round(2).reset_index()
stats_data_2
```

Из таблицы видно, что среднее квадратическое отклонение различно для данных выборок.

Рассмотрим гипотезы:


Нулевая:

H_0: 	Средние пользовательские рейтинги жанров Action   = 	Средние пользовательские рейтинги жанров Sports

`
Альтернативная:

H_a: Средние пользовательские рейтинги жанров Action  ≠ Средние пользовательские рейтинги жанров Sports 


Так как мы рассматриваем гипотезу о равенстве средних, то будем использовать метод для проверки гипотезы scipy.stats.ttest_ind (array1, array2, equal_var)

```python
# подготовим выборки
rating_action = data[(data['genre'] == "Action") & (data['year_of_release'] >=2013) & (data['user_score'] >=0)]['user_score']
rating_sports = data[(data['genre'] == "Sports") & (data['year_of_release'] >=2013) & (data['user_score'] >=0)]['user_score']

# вызов метода для проверки гипотезы
results = st.ttest_ind(rating_action, rating_sports, equal_var=False) 

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вывод:

 Получив p-value, мы отвергли Нулевую гипотезу. Таким образом, вероятность получить одинакове средние рейнтинги по двум жанрам практически нулевая.
